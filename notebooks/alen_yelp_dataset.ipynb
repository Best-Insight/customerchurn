```python
import tensorflow as tf

data_folder = '../raw_data/test'
```

```python
def data_generator(folder):
    """
    get data frome given folder
    Returns:
        iterate data
    """
    files = [file for file in os.listdir(folder)]
    for file in files:
        file_path = os.path.join(folder, file)
        # read one file
        rows = pd.read_csv(file_path)
        for i in range(512):
            x = rows.review.iloc[i]
            if rows.star.iloc[i] >=4:
                y = 1
            else:
                y = 0
            yield x, y
            
dataset = tf.data.Dataset.from_generator(lambda: data_generator(data_folder), output_types=(tf.string, tf.bool)).batch(256).repeat()

```
